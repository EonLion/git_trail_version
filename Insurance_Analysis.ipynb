```python
# How do different demographics (like age, gender, or region) affect premiums
```

```python
sns.lineplot(x='age',y='charges',data=insurance_data)
```

```python
male_count = insurance_data[insurance_data['sex'] == 'male'].shape[0]
female_count = insurance_data[insurance_data['sex'] == 'female'].shape[0]
```

```python
print(f'Male:{male_count}')
print()
print(f'Female:{female_count}')
```

```python
sns.barplot(x='sex',y='charges',data=insurance_data)
```

```python
sns.barplot(x='region',y='charges',data=insurance_data)
```

```python
smoker_data = insurance_data[insurance_data['smoker'] == 'yes']
numeric_smoker_data = smoker_data.select_dtypes(include=['number'])
smoker_charges_corr = numeric_smoker_data.corr()['charges']
print(smoker_charges_corr)
```
